```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = np.arange(10)
plt.plot(data);
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, (3, 4))

ax1.hist(np.random.standard_normal(100), bins=20, color='black', alpha=0.7);
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.standard_exponential(30));
ax3.plot((np.random.standard_normal(50) / 3).cumsum(), color='black', linestyle='dashed', marker='o');
```
